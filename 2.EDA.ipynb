```python
#importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

HOMICIDIOS

```python
#cargar los archivos
df_hom = pd.read_csv('datasets\df_homicidios.csv')
df_hom_vic = pd.read_csv('datasets\df_vic_hom.csv')
```

Usar el primer df_hom

```python
# Cambio de tipo a la columna hora
df_hom['HORA'] = df_hom['HORA'].fillna(0).astype('int')

# cambio a formato categ√≥rico
df_hom['TIPO_DE_CALLE'] = df_hom['TIPO_DE_CALLE'].astype('category')
df_hom['COMUNA'] = df_hom['COMUNA'].astype('category')
df_hom['VEHICULO_ACUSADO'] = df_hom['VEHICULO_ACUSADO'].astype('category')

# Fecha
df_hom['FECHA'] = df_hom['FECHA'].astype('datetime64[ns]')


# cambio a float los valores de latitud y longitud
df_hom['LONGITUD'] = df_hom['LONGITUD'].astype('float')
df_hom['LATITUD'] = df_hom['LATITUD'].astype('float')

df_hom.info()
```

Grafico de las variables

```python
#definir columnas para sacar los graficos
columnas1=["TIPO_DE_CALLE","COMUNA","VEHICULO_ACUSADO","N_VICTIMAS"]
```

```python
#ver las frecuencias de los datos categoricos de la primera tabla df1
for col in columnas1:
    df_hom[col].value_counts().plot(kind='bar',title="Distribucion de "+str(col),)
    plt.show()
```

subplots de la fecha, a√±o, mes, dia

```python
# SUBPLOTS DE LOS A√ëOS, MESES, DIAS Y HORAS DE LOS HOMICIDIOS

# Crear el gr√°fico
plt.figure(figsize=(10, 7))

# Subplot 1: Count plot para los meses de la columna FECHA
plt.subplot(2, 2, 1)  
sns.countplot(x=df_hom['FECHA'].dt.year)
plt.title('Homicidios por a√±o')
plt.xlabel('A√±o')
plt.ylabel('Cantidad')

# Subplot 2: Count plot para los d√≠as de la semana de la columna FECHA
plt.subplot(2, 2, 2)  
sns.countplot(x=df_hom['FECHA'].dt.month )
plt.title('Homicidios por mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad')

# Subplot 3: Count plot para las horas de la columna HORA
plt.subplot(2, 2, 3)  
sns.countplot(x=df_hom['FECHA'].dt.dayofweek)
plt.title('Homicidios por d√≠a de la semana')
plt.xlabel('D√≠a de la semana')
plt.ylabel('Cantidad')

# Subplot 4: Count plot para los a√±os de la columna FECHA
plt.subplot(2, 2, 4)  
sns.countplot(x=df_hom['HORA'])
plt.title('Homicidios por hora')
plt.xlabel('Hora')
plt.ylabel('Cantidad')

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```

LESIONES

```python
df_les = pd.read_csv('datasets\df_lesiones.csv')
df_les_vic = pd.read_csv('datasets\df_vic_les.csv')
```

Usar el archivo df_lesiones

```python
# Cambio de tipo a la columna hora
df_les['HORA'] = df_les['HORA'].fillna(0).astype('int')

# cambio a formato categ√≥rico
df_les['TIPO_DE_CALLE'] = df_les['TIPO_DE_CALLE'].astype('category')
df_les['COMUNA'] = df_les['COMUNA'].astype('category')
df_les['VEHICULO_ACUSADO'] = df_les['VEHICULO_ACUSADO'].astype('category')

# FECHA
df_les['FECHA'] = df_les['FECHA'].astype('datetime64[ns]')

df_les.info()
```

graficar las variables

```python
#definir columnas para sacar los graficos
columnas2=["TIPO_DE_CALLE","COMUNA","VEHICULO_ACUSADO","N_VICTIMAS"]
#ver las frecuencias de los datos categoricos de la primera tabla df_lesiones
for col in columnas2:
    df_les[col].value_counts().plot(kind='bar',title="Distribucion de "+str(col))
    plt.show()
```

```python
# subplots de los a√±os, meses, dias de las lesiones

# Crear el gr√°fico
plt.figure(figsize=(10, 7))

# Subplot 1: Count plot para los meses de la columna FECHA
plt.subplot(2, 2, 1)
sns.countplot(x=df_les['FECHA'].dt.year)
plt.title('Lesiones por a√±o')
plt.xlabel('A√±o')
plt.ylabel('Cantidad')

# Subplot 2: Count plot para los d√≠as de la semana de la columna FECHA
plt.subplot(2, 2, 2)
sns.countplot(x=df_les['FECHA'].dt.month)
plt.title('Lesiones por mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad')

# Subplot 3: Count plot para las horas de la columna HORA
plt.subplot(2, 2, 3)
sns.countplot(x=df_les['FECHA'].dt.dayofweek)
plt.title('Lesiones por d√≠a de la semana')
plt.xlabel('D√≠a de la semana')
plt.ylabel('Cantidad')

# Subplot 4: Count plot para los a√±os de la columna FECHA
plt.subplot(2, 2, 4)
sns.countplot(x=df_les['HORA'])
plt.title('Lesiones por hora')
plt.xlabel('Hora')
plt.ylabel('Cantidad')

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```
